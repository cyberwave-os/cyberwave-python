### Authenticate with the Cyberwave CLI

Run **`cyberwave login`** so credentials are saved for the Python SDK (needed for **workspace model slugs** / Playground-backed loads).

**Interactive:** the next cell prompts for email/password in supporting notebook frontends.

**Non-interactive:** `cyberwave login --email ... --password ...` (avoid committing notebooks that embed passwords).

You can skip login if you only use **local Ultralytics hub IDs** such as `yolo26n.pt` with no cloud routing.

```python
IMAGE_PATH = "engineer-desk.png"
MODEL_ID = "yolo26n.pt"
```

Then let's grep the CYBERWAVE_API_KEY if you already logged in with the cyberwave-cli `cyberwave login --token YOUR_TOKEN`

```python
import json
from pathlib import Path

path = Path("~/.cyberwave/credentials.json").expanduser()
cw_credentials = json.loads(open(path).read())
cw_api_key = cw_credentials["token"]
# Suppress output of this cell
%env CYBERWAVE_API_KEY=$cw_api_key
```

Or just set the API_KEY directly

```python
%env CYBERWAVE_API_KEY=YOUR_TOKEN
```

```python
from PIL import Image

import cyberwave as cw

image = Image.open(IMAGE_PATH)

model = cw.models.load(MODEL_ID)
pred = model.predict(image)
pred
```
